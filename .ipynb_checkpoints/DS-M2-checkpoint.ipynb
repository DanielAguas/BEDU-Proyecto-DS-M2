### Cambio y traducción de nombres

```python
column_name_mapping = {
    'Car_id' : 'id_Auto' ,
    'Date' : 'Fecha_Compra',
    'Customer Name': 'Cliente',
    'Gender': 'Género',
    'Annual Income': 'Ingreso_Anual',
    'Dealer_Name': 'Agencia',
    'Company': 'Marca',
    'Model': 'Modelo',
    'Engine': 'Motor',
    'Transmission': 'Transmisión',
    'Color': 'Pintura',
    'Price ($)': 'Precio',
    'Body Style': 'Carrocería',
    'Dealer_Region': 'Región'
}
df_1_renamed = df_1.rename(columns = column_name_mapping)
df_1_renamed.head()
```

### Identificación de NaNs por columna y fila

```python
df_1_renamed.isna().sum()
```

```python
df_1_renamed.isna().sum(axis=1)
```

### Llenando el único valor nulo de Cliente

```python
df_1_renamed['Cliente'] = df_1_renamed['Cliente'].fillna('Unknown')
df_1_renamed.isna().sum()
```

```python
df_1_renamed.dtypes
```

### Convirtiendo el tipo de dato 'Fecha_Compra' de object a date

```python
pd.to_datetime(df_1_renamed['Fecha_Compra'])
```

```python
df_1_renamed['Fecha_Compra'] = pd.to_datetime(df_1_renamed['Fecha_Compra'])
df_1_renamed.dtypes
```

### Limpiando strings de Data Frame: Columna Motor, quitando 'DoubleÂ'

```python
df_1_renamed['Motor'] = df_1_renamed['Motor'].str.replace('DoubleÂ','Double')
df_1_renamed
```

### Traducción de las str en las columnas 'Motor', 'Transmisión', 'Pintura' y 'Carrocería'

```python
#Mapa Motor:
tipo_motor = {
    'Double\xa0Overhead Camshaft': 'Doble Árbol de Levas',
    'Overhead Camshaft': 'Árbol de Levas Simple',
}
#Método map para transofrmación de dato:
df_1_renamed['Motor'].map(tipo_motor)
```

```python
#Asentando el nuevo cambio a 'Motor'
df_1_renamed['Motor'] = df_1_renamed['Motor'].map(tipo_motor)
```

```python
df_1_renamed
```

```python
#Mapa Transmisión:
tipo_transmision = {
    'Auto': 'Automática',
    'Manual': 'Manual'
}
#Método map para transofrmación de dato:
df_1_renamed['Transmisión'].map(tipo_transmision)
```

```python
#Asentando el nuevo cambio a 'Transmisión'
df_1_renamed['Transmisión'] = df_1_renamed['Transmisión'].map(tipo_transmision)
df_1_renamed
```

```python
#Mapa Pintura:
color_pintura = {
    'Black': 'Negra',
    'Red': 'Roja',
    'Pale White': 'Blanca'
}
#Método map para transofrmación de dato:
df_1_renamed['Pintura'].map(color_pintura)
```

```python
#Asentando el nuevo cambio a 'Pintura'
df_1_renamed['Pintura'] = df_1_renamed['Pintura'].map(color_pintura)
df_1_renamed
```

```python
#Mapa Carrocería:
tipo_carroceria = {
    'SUV': 'SUV',
    'Passenger': 'Dos puertas', 
    'Hatchback': 'Hatchback',
    'Hardtop': 'Carga',
    'Sedan': 'Sedán'
}
#Método map para transofrmación de dato:
df_1_renamed['Carrocería'].map(tipo_carroceria)
```

```python
#Asentando el nuevo cambio a 'Carrocería'
df_1_renamed['Carrocería'] = df_1_renamed['Carrocería'].map(tipo_carroceria)
df_1_renamed
```

```python
### Traduciendo la Columna 'Género'
#Identificando valores unicos para las 4 colmnas:
df_1_renamed['Género'].unique()
```

```python
#Traduciendo Género:

t_genero = {
    'Male': 'Masculino', 
    'Female': 'Femenino'
}

def traduccion_genero(valor):
    return t_genero.get(valor, valor)

# Aplicar la función utilizando un bucle for
for i in range(len(df_1_renamed)):
    df_1_renamed.loc[i, 'Género'] = traduccion_genero(df_1_renamed.loc[i, 'Género'])

df_1_renamed
```

### Algunos nombres no corresponden con el Género de los clientes, por lo que la información esta invertida ó los clientes eligieron ese nombre

```python
## Sueldo más bajo
df_1_sorted = df_1_renamed.sort_values(by='Ingreso_Anual', ascending=True)
ingreso_mas_bajo = df_1_sorted.iloc[0]
print(ingreso_mas_bajo)
```

### El ingreso anual no limita la compra de un vehiculo

```python
df_1_renamed.groupby('Género').size()
```

### El género masculino tiende a comprar más autos

```python
df_femenino = df_1_renamed[df_1_renamed['Género'] == 'Femenino']
compras_femenino_por_modelo = df_femenino.groupby(['Marca', 'Modelo', 'Carrocería']).size()
compras_femenino_por_modelo = compras_femenino_por_modelo.sort_values(ascending=False)

print(compras_femenino_por_modelo.head(10))
```

### El auto más comprado por el genero femenino es la SUV Acura RL

```python
df_masculino = df_1_renamed[df_1_renamed['Género'] == 'Masculino']
compras_masculino_por_modelo = df_masculino.groupby(['Marca', 'Modelo', 'Carrocería']).size()
compras_masculino_por_modelo = compras_masculino_por_modelo.sort_values(ascending=False)
print(compras_masculino_por_modelo.head(10))
```

### El auto más comprado por el genero masculino es el Oldsmobile Silhouette

```python
# Variación del precio de los autos en un intervalo de 2 años
df_1_renamed['Año'] = df_1_renamed['Fecha_Compra'].dt.year
df_ultimos_dos_anios = df_1_renamed[df_1_renamed['Año'] >= (df_1_renamed['Año'].max() - 1)]
variacion_precios = df_ultimos_dos_anios.groupby(['Marca', 'Modelo'])['Precio'].agg(lambda x: x.iloc[-1] - x.iloc[0]).reset_index()
print(variacion_precios.head(10))
```

### Algunos modelos no han variado en precio, otros sufrieron de una gran devaluación, probablemente por el mercado de segunda mano

```python
# Agencia con más ventas
agrupacion_por_agencia = df_1_renamed.groupby('Agencia').size()
agrupacion_por_agencia_ordenada = agrupacion_por_agencia.sort_values(ascending=False)
print(agrupacion_por_agencia_ordenada. head())
```

### La agencia que más ventas concreto fue 'Progressive Shippers Cooperative Association No'

```python
#Gráfica Pastel de las agencias con más ventas:
Ventas = [1318, 1313, 1253, 1251, 1249]
Agencia = ['Progressive Shippers Cooperative Association No', 'Rabun Used Car Sales', 'Race Car Help', 'Saab-Belle Dodge', 'Tri-State Mack Inc']
plt.pie(Ventas, labels = Agencia)
plt.title('Top 5 Agencias con más ventas')
plt.show()
```

```python
# Armadora de autos que más vendió
ventas_por_armadora = df_1_renamed.groupby('Marca').size()
resultados = pd.DataFrame({'Armadora': ventas_por_armadora.index, 'Ventas': ventas_por_armadora.values})
armadora_mas_ventas = resultados.loc[resultados['Ventas'].idxmax()]
print(armadora_mas_ventas)
```
